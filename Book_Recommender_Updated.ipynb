```python
#merge ratings and books on "ISBN" column
ratings_with_book_names =  ratings.merge(books , on = 'ISBN')
```

```python
ratings_with_book_names
```

```python
no_of_ratings_df = ratings_with_book_names.groupby('Book-Title').count()['Book-Rating'].reset_index()
no_of_ratings_df.rename(columns = {'Book-Rating':'num_of_ratings'},inplace = True)
no_of_ratings_df
```

```python
average_rating_df = ratings_with_book_names.groupby('Book-Title').mean()['Book-Rating'].reset_index()
average_rating_df.rename(columns = {'Book-Rating':'average_rating'},inplace = True)
average_rating_df
```

```python
popularity_df = no_of_ratings_df.merge(average_rating_df,on = 'Book-Title')
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['num_of_ratings']>=250].sort_values('average_rating',ascending = False).head(20)
```

```python
popularity_df = popularity_df.merge(books , on= 'Book-Title').drop_duplicates('Book-Title')
```

```python
popularity_df = popularity_df.merge(books , on= 'Book-Title').drop_duplicates('Book-Title')[['Book-Title' , 'Book-Author' , 'Image-URL-M' , 'num_of_ratings' , 'average_rating']]
```

## Collaborative filtering based recommender system

```python
x = ratings_with_book_names.groupby('User-ID').count()['Book-Rating'] > 200

users_who_give_ratings = x[x].index
```

```python
ratings_with_book_names
```

```python
flt_rating = ratings_with_book_names[ratings_with_book_names['User-ID'].isin(users_who_give_ratings)]
```

```python
y = flt_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_book = y[y].index
```

```python
fnl_ratings = flt_rating[flt_rating['Book-Title'].isin(famous_book)]
```

```python
fnl_ratings.drop_duplicates()
```

```python
pvt = fnl_ratings.pivot_table(index='Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pvt.fillna(0,inplace = True)
```

```python
pvt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pvt)
```

```python
similarity_score[0]
```

```python
similarity_score.shape
```

```python
def recommender(book_name):
    #index fetch
    index = np.where(pvt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse = True
      )[1:5]
    #distances = similarity_scores[index]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pvt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommender('1984')
```
